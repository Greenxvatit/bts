```python
def cv(df1,cycle_id,clr,label,grad=False):
    voltage = []
    current = []
    for i in range(0, len(df1)):
        if df1['cycle number'][i] == cycle_id:
            voltage.append(df1['control_V'][i])
            current.append(df1['<I>'][i])
    plt.plot(np.array(voltage[:(len(voltage)-25)]),1000*np.array(current[:(len(voltage)-25)])/(0.78**2 * np.pi),color=f'{clr}', label=f'{label}')
```

```python
fig,ax = plt.subplots(figsize=(7, 5),dpi=400)
plt.xlabel("Potential (V vs Li$^+$/Li)",fontsize=18)
plt.ylabel("i($\mu$A/$cm^2$) ",fontsize=18)
cv(dmp12,1,'k','1,2-DMP:FEC(9:1) 1M LiBF$_{4}$')
cv(dmp13,1,'r','1,3-DMP:FEC(9:1) 1M LiBF$_{4}$')
cv(dmb,1,'b','DMB:FEC(9:1) 1M LiBF$_{4}$')
cv(dig,1,'navy','Diglyme:FEC(9:1) 1M LiBF$_{4}$')
cv(ec_dmc,1,'g','EC:DMC(1:1) 1M LiPF$_{6}$')
plt.ylim(0,6)
plt.xlim(2.7,5.5)
plt.legend(loc="upper left",frameon=False)
#  интервал основных делений по оси x:
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
#  интервал вспомогательных делений по оси x:
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.2))
#  интервал основных делений по оси y:
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
#  интервал вспомогательных делений по оси y:
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.5))
plt.savefig('lsv_ethers.png')
# plt.savefig('lsv 75ru.pdf',format='pdf')
```
